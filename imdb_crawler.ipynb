```python
def get_movie_info(movie_keyword, top_nums_movies):
    """
    movie_keyword is a string ' ' or " " are needed
    top_nums_movies is an integer 
    This function returns movie's 1) name 2) rating 3) how long the movie is 
    4) gerne 5) cast 6) county & released date in a json format
    It returns the results also exports as a "movie_keyword.json" file
    """
    from selenium import webdriver
    from webdriver_manager.chrome import ChromeDriverManager
    import json
    
    driver = webdriver.Chrome(ChromeDriverManager().install())
    # Jump to imdb home page
    driver.get('https://www.imdb.com/')
    # Enter the keyword to the search bar and hit search
    elem = driver.find_element_by_xpath('//*[(@id = "navbar-query")]')
    elem.send_keys(movie_keyword)
    elem = driver.find_element_by_xpath('//*[(@id = "navbar-submit-button")]')
    elem.click()
    # Jump to the page that shows all searches
    elem = driver.find_element_by_xpath('//*[contains(concat( " ", @class, " " ), concat( " ", "findSection", " " )) and (((count(preceding-sibling::*) + 1) = 5) and parent::*)]//*[contains(concat( " ", @class, " " ), concat( " ", "odd", " " ))]//a')
    elem.click()

    #Build up the path

    movie_info=[]

    for i in range(1,top_nums_movies+1):
        # Combine to make a series of the xpaths
        xpath1 ="//div[@class='lister-item mode-detail']["
        xpath2="]/div[@class='lister-item-content']/h3[@class='lister-item-header']/a"
        xpath = xpath1+str(i)+xpath2
        elem = driver.find_element_by_xpath(xpath)
        elem.click()
        # Get the name
        elem = driver.find_element_by_xpath("//div[@class='title_wrapper']/h1")
        name = elem.text
        # Get the rating 
        elem = driver.find_element_by_xpath('//strong/span')
        rating = float(elem.text)
        # Get the movie's gerne
        elem = driver.find_elements_by_xpath('//*[contains(concat( " ", @class, " " ), concat( " ", "subtext", " " ))]//a')
        gerne = [i.text for i in elem]
        gerne.pop()
        # Get the movie's time
        elem = driver.find_element_by_xpath("//div[@class='subtext']/time")
        movie_time = elem.text
        # Jump to the full cast page
        # Get the movie's cast
        elem = driver.find_element_by_xpath("//div[@class='credit_summary_item'][3]/a[4]")
        elem.click()
        elem = driver.find_elements_by_css_selector('.primary_photo+ td a')
        cast = [i.text for i in elem]
        # Go back to the last page
        driver.back()
        # Jump to the release country and date page
        # Get the movie's release countries and dates
        elem = driver.find_element_by_xpath("//div[@class='subtext']/a[4]")
        elem.click()
        elem = driver.find_elements_by_css_selector('.release-date-item__country-name a')
        countries = [i.text for i in elem]
        elem = driver.find_elements_by_css_selector('.release-date-item__date')
        release_date = [i.text for i in elem]
        # Zip countries and dates as a dictionary and rename it as release_date
        release_date = dict(zip(countries, release_date))
        # Combine all the info into a movie dictionary
        movie = {'movie':name, 'Rating': rating, 'Times': movie_time, 'Gerne':gerne, 'Cast':cast, 'County': release_date}
        movie_info.append(movie)
        driver.back()
        driver.back()
        
        filename = movie_keyword+'.json'
        with open(filename, 'w') as json_file:  
            json.dump(movie_info,json_file)
    driver.close()
    return movie_info
```
